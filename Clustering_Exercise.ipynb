```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

#### PART 1 - Dummy data

```python
# Load the clusters csv into a dataframe
data_df = pd.read_csv("sample_data/clusters.csv")
```

```python
plt.scatter(data_df["x1"], data_df["x2"])
plt.show()
```

```python
# Run k-means over the clusters dataset with 3 clusters
def get_kmeans(data_df, n_clusters):
  km = KMeans(n_clusters)
  km = km.fit(data_df)

  return km

km = get_kmeans(data_df, 3)
```

```python
# What is the inertia (dissimilarity) of this clustering?
km.inertia_
```

```python
# Plot the learned clusters which each cluster being a different color (create a function for this)
def plot_clusters(data_df, km):
  plt.scatter(data_df["x1"], data_df["x2"], c = km.labels_)
  plt.show()

plot_clusters(data_df, km)
```
